The aim of this task is to create a Decision tree algorithm and to visualize it graphically & to feed a new data to it so that it would be able to predict the class accurately.

```python
#importing python packages
import numpy as np
import pandas as pd
import seaborn as sns
sns.set_palette('husl')
import matplotlib.pyplot as plt
%matplotlib inline
import sklearn.datasets as datasets
from sklearn.model_selection import train_test_split
# The Model
from sklearn.tree import DecisionTreeClassifier
```

```python
#importing dataset from locallink
# Loading the iris dataset
iris=datasets.load_iris()
# Forming the iris dataframe
df=pd.DataFrame(iris.data, columns=iris.feature_names)
print(df.head(5))

y=iris.target
print(y)
```

# Exploring data

```python
df.describe()
```

There are 150 rows & 6 columns  in the dataset.There is no null value in the data.There are 50 observations of each species (setosa, versicolor, virginica).

# Visualizing Data

```python
corr=df.corr()
sns.heatmap(corr)
```

 A heatmap is drawn to find the correlation among the different features.

# Splitting the dataset & Fitting the model

```python
# Splitting the dataset into train and test set
X_train,X_test,y_train,y_test=train_test_split(df,y,test_size=0.2,random_state=42)
```

```python
dtree=DecisionTreeClassifier()# Creating an object of the DecisionTreeClassifier
dtree.fit(X_train,y_train)# Fitting the training data
print("The model has been trained.")
```

```python
y_pred=dtree.predict(X_test)# Predicting on X_test
print("The predicted values are:\n",y_pred)
print("\nThe actual values are:\n",y_test)
```

# Visualizing the Decision tree

```python
from sklearn import tree
import graphviz
```

```python

# Method 1: Using plot_tree
fig = plt.figure(figsize=(25,20))
tree.plot_tree(dtree, 
               feature_names=iris.feature_names,  
                class_names=iris.target_names,
                filled=True);
```
